```python
#Twitter URL
twitter_url = 'https://twitter.com/marswxreport?lang=en'
```

```python
Twitter = urllib.request.urlopen(twitter_url).read()
```

```python
Twitter_Soup = BeautifulSoup(Twitter, 'html.parser')
Twitter_Soup
```

```python
Weather_Tweets= Twitter_Soup.find('div', attrs={"class": "tweet", "data-name": "Mars Weather"})

Weather_Tweets
```

```python
# Next, search within the tweet for the p tag or span tag containing the tweet text
# As Twitter is frequently making changes the try/except will identify the tweet
# text usings a regular expression pattern that includes the string 'sol' if there
# is no p tag with a class of 'tweet-text'
try:
    weather= Weather_Tweets.find("p", "tweet-text").get_text()
    weather
except AttributeError:
    pattern = re.compile(r'sol')
    weather = Twitter_Soup.find('span', text=pattern).text
    weather
weather
```
